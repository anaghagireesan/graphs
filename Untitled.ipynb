```python
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
import matplotlib.patheffects as path_effects

headers = ['Steps','Latency']
df1 = pd.read_csv('softlatency.csv',delimiter = ", ")
df2 =pd.read_csv('hardlatency.csv',delimiter = ", ")
plt.grid(True, color="#93a1a1", alpha=0.3)
plt.plot(df1['Steps'],df1['Average Latency'],'#E1D5E7',label='Soft Latency',path_effects=[path_effects.Stroke(linewidth=2, foreground='#9673A6'),
                       path_effects.Normal()])
plt.plot(df2['Steps'],df2['Average Latency'],'#FFF2CC',label='Hard Latency',path_effects=[path_effects.Stroke(linewidth=2, foreground='#D6B656'),
                       path_effects.Normal()])
plt.legend(loc='upper right')
plt.xlabel('Steps')
plt.ylabel('Average Latency')
plt.savefig('latency_supervised.png')
```

```python
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
import matplotlib.patheffects as path_effects

headers = ['Steps','Average Reward']
df1 = pd.read_csv('reward.csv',delimiter = ", ")
df2 =pd.read_csv('random.csv',delimiter = ";")
df3 =pd.read_csv('worse.csv',delimiter = ";")
df4 =pd.read_csv('greedy.csv',delimiter = ";")
plt.grid(True, color="#93a1a1", alpha=0.3)
plt.plot(df1['Steps'],df1['Average Reward'],'#E1D5E7',label='My method',path_effects=[path_effects.Stroke(linewidth=2, foreground='#9673A6'),
                       path_effects.Normal()])
plt.plot(df2['Steps'],df2['Average Reward'],'#FFF2CC',label='Random',path_effects=[path_effects.Stroke(linewidth=2, foreground='#D6B656'),
                       path_effects.Normal()])
plt.plot(df3['Steps'],df3['Average Reward'],'#FFCE9F',label='No migration',path_effects=[path_effects.Stroke(linewidth=2, foreground='#D6B656'),
                       path_effects.Normal()])
plt.plot(df4['Steps'],df4['Average Reward'],'#CCCCCC',label='Greedy',path_effects=[path_effects.Stroke(linewidth=2, foreground='#D6B656'),
                       path_effects.Normal()])
plt.legend(loc='upper right')
plt.xlabel('Steps')
plt.ylabel('Average Reward')
plt.savefig('Reward_reinforce.png')
```
